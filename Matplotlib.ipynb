```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
# fontdict and loc is for text changes loc(left,center,right)
plt.title("Sports Watch Data", fontdict = {'family':'Serif'},loc='left')
plt.xlabel("Average Pulse", fontdict = {'family':'monospace'})
plt.ylabel("Calorie Burnage", fontdict = {'family':'monospace'})
plt.plot(x1,y1)
plt.show()
```

```python
x=np.array([10,20,30])
y=np.array([1,5,6])
plt.title('Second Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# grid for lines
plt.grid()
plt.plot(x,y)
plt.show()
```

```python
x=np.array([10,20,30])
y=np.array([5,15,25])
plt.title('Second Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(axis='x',linestyle='-.',color='purple')
plt.grid(axis='y',linestyle='--',color='purple',linewidth=2)
plt.plot(x,y,marker='o',markeredgecolor='red',markerfacecolor='green',markersize=10)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import seaborn as sns

tips = sns.load_dataset("tips")

sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
# customize the scatter plot
sns.relplot(x="total_bill", y="tip", hue="sex", size="size", sizes=(50, 200), data=tips)
# relplot-scatterplot,lineplot
# add labels and title
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Relationship between Total Bill and Tip")
# display the plot
plt.show()
```

```python
a=sns.load_dataset('flights')
sns.displot(x='year',y='month',hue='passengers',data=a)
# displot-histplot,kdeplot,ecdfplot,rugplot
```

```python
a=sns.load_dataset('flights')
sns.catplot(x='year',y='month',hue='passengers',data=a)
# catplot-striplot,swarmmplot,boxplot,viloinplot,pointplot,barplot
```
